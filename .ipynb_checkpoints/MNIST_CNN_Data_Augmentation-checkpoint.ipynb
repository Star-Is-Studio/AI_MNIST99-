```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.examples.tutorials.mnist import input_data
import random
np.random.seed(80013493)
tf.set_random_seed(80013493)
mnist = input_data.read_data_sets("/tmp/data/", one_hot=True)
```

```python
#data 1000개 추출 =>mnist train의 총개수 : 55000개
num_data = 55000
#augment수 : 1000*0.1 = 100 10퍼센트만 확장시킴
num_augment = 55000
#total개수
num_total = num_data + num_augment
batch_xs, batch_ts = mnist.train.images, mnist.train.labels
test_xs, test_ts = mnist.test.images,mnist.test.labels
```

```python
datagen = ImageDataGenerator(
        rotation_range=5,  
        zoom_range = 0.1,  
        width_shift_range=0.2, 
        height_shift_range=0.2)
```

```python
batch_xs = batch_xs.reshape(-1,28,28,1)
```

```python
t_data = []
for i in range(num_data):
  t_data = np.append(t_data,np.argmax(batch_ts[i]))
```

```python
a = batch_xs
b = t_data
c,d = datagen.flow(batch_xs,t_data,batch_size=num_augment).next()
c = c.reshape(-1,28,28,1)
for i in range(num_augment):
  t_data = np.append(t_data,np.argmax(batch_ts[i]))
#기존배열 뒤에 붙임
a = np.append(a,c)
a = a.reshape(-1,28,28,1)
b = np.append(b,d)
```

```python
t_train = np.zeros((num_total,10))
tmp = np.zeros((num_total,10))
for i in range(num_total):
  j = 0
  for j in range(9):
    if b[i] == j:
      tmp[i][j] = 1
    else:
      tmp[i][j] = 0
```

```python
#형태 맞추기, 변수명 명확히하기
x_train = a.reshape(-1,784)
t_train = tmp
```

```python
#데이터의 길이 확인
print('batch_xs',len(batch_xs))
print('x_train',len(x_train))
print('batch_ts',len(batch_ts))
print('t_train',len(t_train))
```

```python
#확장데이터 확인 1001~1100
fig = plt.figure(figsize=(8,15))
for i in range(100):
  c=0
  index = i+55000
  actual = np.argmax(t_train[index])
  subplot = fig.add_subplot(10,10,i+1+c)
  subplot.set_xticks([])
  subplot.set_yticks([])
  subplot.set_title('%d, %d' %(index, actual))
  subplot.imshow(x_train[index].reshape((28,28)), vmin=0,vmax=1, cmap=plt.cm.gray_r, interpolation="nearest")
  c = i%10
```

```python
#첫번째 필터의 개수는 32개, 두번째는 64개, 마지막은 7*7*64를 주었다.
num_filters1 = 32
num_filters2 = 64
num_units = 7*7*num_filters2

x = tf.placeholder(tf.float32, [None, 784])
x_image = tf.reshape(x, [-1,28,28,1])

W_conv1 = tf.get_variable("w_conv1",[5,5,1,num_filters1],initializer=tf.contrib.layers.xavier_initializer())
h_conv1 = tf.nn.conv2d(x_image, W_conv1, strides=[1,1,1,1], padding='SAME')
b_conv1 = tf.Variable(tf.constant(0.1, shape=[num_filters1]))
h_conv1_cutoff = tf.nn.relu(h_conv1 + b_conv1)

h_pool1 = tf.nn.max_pool(h_conv1_cutoff, ksize=[1,2,2,1], strides=[1,2,2,1], padding='SAME')

W_conv2 = tf.get_variable("w_conv2",[5,5,num_filters1, num_filters2],initializer=tf.contrib.layers.xavier_initializer())
h_conv2 = tf.nn.conv2d(h_pool1, W_conv2, strides=[1,1,1,1], padding='SAME')
b_conv2 = tf.Variable(tf.constant(0.1, shape=[num_filters2]))
h_conv2_cutoff = tf.nn.relu(h_conv2 + b_conv2)

h_pool2 = tf.nn.max_pool(h_conv2_cutoff, ksize=[1,2,2,1], strides=[1,2,2,1], padding='SAME')
h_pool2_flat = tf.reshape(h_pool2, [-1, 7*7*num_filters2])

w1 = tf.Variable(tf.truncated_normal([num_units, num_units]))
b1 = tf.Variable(tf.constant(0.1, shape=[num_units]))
L2 = tf.nn.relu(tf.matmul(h_pool2_flat, w1) + b1)

keep_prob = tf.placeholder(tf.float32)
L2_drop = tf.nn.dropout(L2, keep_prob)

w2 = tf.Variable(tf.zeros([num_units, 10]))
b2 = tf.Variable(tf.zeros([10]))
p = tf.nn.softmax(tf.matmul(L2_drop, w2) + b2)

t = tf.placeholder(tf.float32, [None, 10])
# define cost/loss & optimizer
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=p, labels=t))
optimizer = tf.train.AdamOptimizer(learning_rate=0.001).minimize(loss)
correct_prediction = tf.equal(tf.argmax(p, 1), tf.argmax(t, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
sess = tf.InteractiveSession()
sess.run(tf.global_variables_initializer())
```

```python
#num_total = 총개수
i = 0
tmp = 0
for _ in range(7000):
  i += 1
  sess.run(optimizer, feed_dict={x:x_train[tmp:tmp+5000], t:t_train[tmp:tmp+5000],keep_prob:0.5})
  if i % 500 == 0:
    loss_vals, acc_vals = [], []
    for c in range(4):
      start = int(len(mnist.test.labels) / 4 * c)
      end = int(len(mnist.test.labels) / 4 * (c + 1))
      loss_val, acc_val = sess.run([loss, accuracy], feed_dict={x:mnist.test.images[start:end], t:mnist.test.labels[start:end], keep_prob:1.0})
      loss_vals.append(loss_val)
      acc_vals.append(acc_val)
    loss_val = np.sum(loss_vals)
    acc_val = np.mean(acc_vals)
    tmp=+5000
    if tmp <= 105000:
      tmp = 0
    print('Step: %d, Loss: %f, Accuracy: %f' %(i, loss_val, acc_val))
```
